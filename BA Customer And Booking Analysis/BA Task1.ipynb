```python
#To perform sentiment analysis we will install textblob
!pip install textblob
```

```python
from textblob import TextBlob
```

```python
#Lets showcase how TextBlob works
#We have found the polarity of the first entry in Review title to be 0.7, this means
#it is positive sentiment.
TextBlob(dfc7.iloc[0,1]).sentiment.polarity
```

```python
#We will now add a polarity column which applies above to each entry in comment_text
polarity = []
for review in dfc7['Review Title']:
    try:
        polarity.append(TextBlob(review).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
dfc7['Polarity']= polarity
dfc7
```

```python
#Want to get a word cloud so I will get some filters
filter1 = dfc7['Polarity']==1
pos_reviews = dfc7[filter1]
filter2= dfc7['Polarity']==-1
neg_reviews = dfc7[filter2]
```

```python
#install wordcloud package
!pip install wordcloud
```

```python
#STOPWORD filters out meaningless words
from wordcloud import WordCloud, STOPWORDS
```

```python
total_reviews_pos = ' '.join(pos_reviews['Review Title'])
```

```python
wordcloud_pos = WordCloud(stopwords = set(STOPWORDS)).generate(total_reviews_pos)
```

```python
plt.imshow(wordcloud_pos);
plt.axis('off');
```

```python
total_reviews_neg = ' '.join(neg_reviews['Review Title'])
```

```python
wordcloud_neg = WordCloud(stopwords = set(STOPWORDS)).generate(total_reviews_neg)
```

```python
plt.imshow(wordcloud_neg);
plt.axis('off');
```

```python
p =len(dfc7[dfc7['Polarity']>0])
```

```python
n = len(dfc7[dfc7['Polarity']==0])
```

```python
neg = len(dfc7[dfc7['Polarity']<0])
```

```python
# Data
labels = ['Positive', 'Neutral', 'Negative']
sizes = [p, n, neg]  # Values should sum up to 100 for a pie chart

# Create a pie chart
plt.figure(figsize=(5, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Title
plt.title('Polarity')

# Show the plot
plt.show()
```
